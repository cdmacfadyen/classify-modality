```python

import numpy as np
import matplotlib.pyplot as plt
# DATA
rescaled_train_accuracies = {
    "ResNet18"  : 100,
    "ResNet50"  : 100,
    "VGG16"     : 99,
}
rescaled_validate_accuracies = {
    "ResNet18"  : 98,
    "ResNet50"  : 98,
    "VGG16"     : 99,
}
rescaled_test_accuracies = {
    "ResNet18"  : 95,
    "ResNet50"  : 96,
    "VGG16"     : 96,
}


rescaled_test_balanced_accuracies = {
    "ResNet18"  : 84,
    "ResNet50"  : 87,
    "VGG16"     : 85,
}

plt.rcdefaults()
plt.rcParams.update(
    {"font.size":26,
    "figure.dpi":300,
    })

width = 0.2
x_ticks = np.arange(len(rescaled_train_accuracies))

fig, ax = plt.subplots(figsize=(12,8))
ax.bar(x_ticks - width, rescaled_train_accuracies.values(), width, label = "Train Set Accuracy")
ax.bar(x_ticks, rescaled_validate_accuracies.values(), width, label = "Validate Set Accuracy")
ax.bar(x_ticks + width, rescaled_test_accuracies.values(), width, label = "Test Set Accuracy")
ax.set_xticks(x_ticks)
ax.set_xticklabels(rescaled_train_accuracies.keys())
ax.legend(bbox_to_anchor = (1.22, 1.0),loc="upper center")
ax.set_ylabel("Accuracy (%)")
ax.set_xlabel("Model Name")
fig.suptitle("Comparing Accuracy of Models on the Train, Validate and Test Sets", wrap=True)
plt.savefig("train-validate-test-rescaled.png", bbox_inches="tight")
```

```python
previous_accuracies = {
    "ResNet18" : 52,
    "ResNet50" : 36,
    "VGG16" : 42,
}

width=0.2
fig, ax = plt.subplots(figsize=(12, 8))
x_ticks = np.arange(len(previous_accuracies))
x_labels = previous_accuracies.keys()
ax.bar( x_ticks + width / 2, previous_accuracies.values(), width, label="Rescaling by Datatype Max")
ax.bar( x_ticks - width / 2, rescaled_test_accuracies.values(), width, label="Rescaling by Image Max")
ax.set_xticks(x_ticks)
ax.set_xticklabels(x_labels)
ax.set_xlabel("Model Name")
ax.set_ylabel("Accuracy (%)")
ax.legend()
ax.legend(bbox_to_anchor = (1.22, 1.0),loc="upper center")

# ax.set_title("Accuracy of Different Feature Scaling Techniques on Test Set", wrap=True)
fig.suptitle("Accuracy of Different Feature Scaling Techniques on Test Set", wrap=True)
plt.savefig("rescaled-compare.png", bbox_inches="tight")
```

```python
val_accs_rescaled = [95, 97, 98, 98, 98]
val_accs_previous = [67, 71, 64, 68, 68]

fig, ax = plt.subplots(figsize=(12, 8))
x = np.arange(len(val_accs_previous))
ax.plot(x, val_accs_previous, label = "Rescale by Datatype Max")
ax.plot(x, val_accs_rescaled, label = "Rescale by Image Max")
ax.set_ylim(40, 100)
ax.set_ylabel("Accuracy at end of epoch (%)")
ax.set_xlabel("Epoch")
fig.suptitle("Validation Set Accuracy every Epoch for ResNet50", wrap=True)
ax.set_title(" Trained with Two Rescaling Techniques", wrap=True)
ax.legend(loc="lower right")

plt.savefig("val-acc.png", bbox_inches="tight")
```
