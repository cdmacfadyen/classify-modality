```python
import numpy as np
import matplotlib.pyplot as plt
# Accuracy

plt.rcParams.update({"font.size":26})

accuracies = {
    "DenseNet" : 78,
    "ResNet18" : 76,
    "ResNet34" : 52,
    "ResNet50" : 68,
    "VGG16" : 85,
    "MNASNet" : 71
}

train_accuracies = {
    "DenseNet" : 99,
    "ResNet18" : 99,
    "ResNet34" : 98,
    "ResNet50" : 99,
    "VGG16" : 99,
    "MNASNet" : 99
}
width = 0.4
x_labels = accuracies.keys()
x = np.arange(0, len(x_labels))
y = accuracies.values()

fig, ax = plt.subplots(figsize=(16,12))
ax.bar(x - width / 2, train_accuracies.values(), width, label="Train Set Accuracy")
ax.bar(x + width / 2, y, width, label="Validate Set Accuracy")
ax.set_xticks(x)
ax.set_xticklabels(x_labels)
ax.legend()
ax.set_ylabel("Accuracy (%)")
ax.set_xlabel("Model Architecure")
ax.set_title("Accuracy of Different Models after Five Training Epochs", wrap=True)
plt.savefig("./model-accs.png")
```

```python
# Regularisation experiment results
val_accuracies = {
    "0" : 87,
    "0.005" : 81,
    "0.01" : 65,
    "0.015" : 25,
    "0.02" : 25,
}

train_accuracies = {
    "0" : 99,
    "0.005" : 98,
    "0.01" : 97,
    "0.015" : 25,
    "0.02" : 25,
}

width = 0.4

x_labels = val_accuracies.keys()
x = np.arange(len(val_accuracies))
val_y = val_accuracies.values()
train_y = train_accuracies.values()

fig, ax = plt.subplots(figsize=(16,12))
ax.bar(x - width / 2, train_y, width, label="Train Accuracy")
ax.bar(x + width / 2, val_y, width, label="Validate Accuracy")
ax.set_xticks(x)
ax.legend()
ax.set_xticklabels(x_labels)
ax.set_ylabel("Accuracy (%)")
ax.set_title("Accuracy of VGG16 with Varying Amounts of L2 Regularisation", wrap=True)
# fig.suptitle("Accuracy of VGG16 with Varying Amounts of L2 Regularisation", wrap=True)
plt.savefig("./l2-accs.png")
```
